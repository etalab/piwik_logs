```python
import pickle
import numpy as np

import pandas as pd
```

# Load indexes etc

```python
dataset_info = pd.read_csv('datasets-2017-06-21-19-43.csv', sep=';')
datasets = list(dataset_info.id)
datasets_set = set(datasets)
datasets_index = {dataset: i for i, dataset in enumerate(datasets)}
n_datasets = len(datasets)
id2slug = {r[1]: r[3] for r in dataset_info.itertuples()}
slug2id = {r[3]: r[1] for r in dataset_info.itertuples()}

keywords = pickle.load(open('keywords_list.pickle', 'rb'))
keywords_set = set(keywords)
keywords_index = {keyword: i for i, keyword in enumerate(keywords)}
n_keywords = len(keywords)
with open('keyword_count.pickle', 'rb') as f:
    keyword_count = pickle.load(f)
```

# Find the most consulted dataset for the most searched keywords

```python
most_searched_keyword_indexes = np.argsort(-keyword_count)
```

```python
adj_keyword_dataset = pickle.load(open('keyword_dataset_adjacency_list.pickle', 'rb'))
```

```python
def pprint_keyword(keyword_index):
    print(keywords[keyword_index].upper())

    sorted_datasets = sorted(adj_keyword_dataset[keyword_index].items(), key = lambda x: -x[1])
    for dataset_index, weight in sorted_datasets[:10]:
        ident = datasets[dataset_index]
        slug = id2slug[ident]
        print("{}\t{}".format(weight, slug))
    print()
```

```python
for keyword_index in most_searched_keyword_indexes[:50]:
    pprint_keyword(keyword_index)
```
